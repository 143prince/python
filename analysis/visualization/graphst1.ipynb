```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

```

```python
df = sns.load_dataset('titanic')
df
```

```python
sns.countplot(df, x='class')
plt.title('Number of passenger in each class')
plt.xlabel('class')
plt.ylabel('Number of passenger')
plt.show()
```

```python
sns.histplot(df, x='age', kde = True, color='red') #kde drw a line in our graph 
plt.title('Age distribution of passengers')
plt.xlabel('Age')
plt.ylabel('Number of passengers')
plt.show()
```

```python
sns.lineplot(df, x='class', y='fare', linestyle='dashed', marker='s', color='green')
plt.title('Fare of each class')
plt.xlabel('Class')
plt.ylabel('Fare')
plt.show()
```
